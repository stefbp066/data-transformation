even within DynamicPrice_Result, the algorithmOutput results differ. but for now we can get out everything else: \
DynamicPriceOption.csv requires:

Provider: in quotes \
OfferId: in quotes \
UniqueOptionId: in quotes \
BestPrice: without quotes \
EnqueuedTimeSP: DD/MM/YYYY (converted to Brasilian timezone - UTC-3)

```python
dpr = df[df['EventName']=='DynamicPrice_Result']
dpr
```

```python
def dynamicprice_option(row):
    time = row['EnqueuedTimeUtc']
    time = pd.to_datetime(time).strftime("%d/%m/%y")
    payload_row = json.loads(row['Payload'])  # converts json string s to Py dictionary
    result = []
    if (payload_row['provider']=='ApplyDynamicPricePerOption'):
        provider = f"\"{payload_row['provider']}\""
        offerid = f"\"{payload_row['offerId']}\""
        for option in (payload_row['algorithmOutput']): # loop because there are several entries of the same structure within the algorithmOutput
            priceoption = {'Provider': provider,
            'OfferId': offerid,
            'UniqueOptionId': f"\"{option['uniqueOptionId']}\"",
            'BestPrice': f"\"{option['bestPrice']}\"",
            'EnqueuedTimeSP': time
            } # collect everything into a dictionary
        result.append(priceoption)
    return result
dpr_list = dpr.apply(dynamicprice_option,axis=1)
dpr_flat = pd.DataFrame([x for xs in dpr_list for x in xs])
dpr_flat['BestPrice'] = dpr_flat['BestPrice'].str.strip().str.replace('"', '').apply(pd.to_numeric)
dpr_flat.to_csv('DynamicPriceOption.csv')
```

```python
def dynamicprice_range(row):
    time = row['EnqueuedTimeUtc']
    time = pd.to_datetime(time).strftime("%d/%m/%y")
    payload_row = json.loads(row['Payload'])  # converts json string s to Py dictionary
    result = []
    if (payload_row['provider']=='ApplyDynamicPriceRange'):
        provider = f"\"{payload_row['provider']}\""
        offer_id = f"\"{payload_row['offerId']}\""
        algorithm_output = payload_row['algorithmOutput']
        price_range = {
        'Provider': provider,
        'OfferId': offer_id,
        'MinGlobal': f"\"{algorithm_output['min_global']}\"",
        'MinRecommended': f"\"{algorithm_output['min_recommended']}\"",
        'MaxRecommended': f"\"{algorithm_output['max_recommended']}\"",
        'DifferenceMinRecommendMinTheory': f"\"{algorithm_output['differenceMinRecommendMinTheory']}\"",
        'EnqueuedTimeSP': time}
        result.append(price_range)
    return result
dpr_list = dpr.apply(dynamicprice_range,axis=1)
dpr_flat = pd.DataFrame([x for xs in dpr_list for x in xs])
col_conv = ['MinGlobal', 'MinRecommended', 'MaxRecommended','DifferenceMinRecommendMinTheory']
for col in col_conv:
    dpr_flat[col] = dpr_flat[col].str.strip().str.replace('"', '').apply(pd.to_numeric) # do loop since dataframes have no attribute str
dpr_flat.to_csv('DynamicPriceRange.csv')
```
